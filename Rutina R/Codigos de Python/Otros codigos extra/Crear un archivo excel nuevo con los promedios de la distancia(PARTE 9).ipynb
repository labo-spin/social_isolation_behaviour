```python
import os
import pandas as pd
import re

# Función para crear un nuevo archivo con la distancia y los nombres de archivo ordenados
def crear_archivo_distancia(carpeta_origen, nombre_archivo_destino):
    # Lista para almacenar los datos
    datos = []

    # Recorrer todos los archivos en la carpeta de origen
    for archivo in os.listdir(carpeta_origen):
        if archivo.endswith('.xlsx'):
            # Usar regex para extraer los números X y B del nombre del archivo
            match = re.search(r'C(\d+)_(\d+)', archivo)
            if match:
                x_value = int(match.group(1))  # Valor de X
                b_value = int(match.group(2))  # Valor de B

                # Cargar el archivo
                df = pd.read_excel(os.path.join(carpeta_origen, archivo))

                # Verificar si la columna "Suma Distancia Euclidiana" existe
                if 'Suma Distancia Euclidiana' in df.columns:
                    # Obtener la columna "Suma Distancia Euclidiana"
                    columna_distancia = df['Suma Distancia Euclidiana']
                    
                    # Encontrar el primer valor numérico en la columna
                    primer_valor_numerico = columna_distancia[pd.to_numeric(columna_distancia, errors='coerce').notnull()].iloc[0]

                    # Agregar el valor y el nombre del archivo a la lista de datos
                    datos.append([primer_valor_numerico, archivo, x_value, b_value])
                else:
                    print(f"La columna 'Suma Distancia Euclidiana' no se encuentra en el archivo {archivo}.")
            else:
                print(f"No se pudo extraer los valores de {archivo}")

    # Crear un DataFrame a partir de los datos recopilados
    df_resultado = pd.DataFrame(datos, columns=['Suma Distancia Euclidiana', 'Nombre del Archivo', 'X', 'B'])

    # Ordenar el DataFrame por X y B
    df_resultado.sort_values(by=['X', 'B'], inplace=True)

    # Seleccionar solo las columnas necesarias para el archivo final
    df_resultado_final = df_resultado[['Suma Distancia Euclidiana', 'Nombre del Archivo']]

    # Guardar el DataFrame en un nuevo archivo Excel
    ruta_destino = os.path.join(carpeta_origen, nombre_archivo_destino)
    df_resultado_final.to_excel(ruta_destino, index=False)

    print(f"Archivo creado: {ruta_destino}")

# Llama a la función con la carpeta de origen y el nombre del nuevo archivo
carpeta_destino = r'C:\Users\Usuario\Desktop\Prueba10'
nombre_archivo_destino = 'Distancia.xlsx'
crear_archivo_distancia(carpeta_destino, nombre_archivo_destino)

```
