```python
##### ESTE CODIGO LO HICIMOS CON NICO 16/10/2024. SE USA PARA LOS VIDEOS DE ESCAPE. SU FUNCIÓN ES LA DE OBTENER LAS VARIABLES CINÉTICAS SOLO PARA LOS TRIALS EN DONDE SE CATALOGARON RESPUESTAS, Y ALREDEDOR DE LAS MISMAS (10 O 20 FRAMES) CALCULAR LAS VARIABLES Y OBTENER UNA DISTRIBUCIÓN DE CADA UNA, PARA LOS AISLADOS Y PARA LOS SOCIALES.

import os
import pandas as pd
import matplotlib.pyplot as plt

Tracking=r"C:\Users\Usuario\Desktop\Nuevo3\Output"
os.chdir(Tracking)

archivos = os.listdir()
nuevos_archivos = []

for archivo in archivos:
    if '.xlsx' in archivo and 'modified_modified_filtrado' in archivo:
        nuevos_archivos.append(archivo)
        
archivos = nuevos_archivos

# Definir la carpeta que contiene los archivos Excel
input_file = r'C:\Users\Usuario\Desktop\Escape nuevo2.xlsx'

# Leer el archivo Excel
df = pd.read_excel(input_file)


resultados = {"A": {}, "S": {}}

contador = 0


eventos_por_condicion = {"A": 0, "S": 0}


for Nombre in archivos:
    #if contador > 300: #este codigo lo que hace es iterar en los primeros 100 archivos nada más.
     #   break          #
    
    contador = contador + 1
    
    datos = Nombre.split("modified_modified_filtrado_")[1].split("DLC")[0].split("_")
    Camada = int(datos[0].split("C")[-1])
    Individuo = int(datos[1])
    Condicion = datos[2]
    Estimulos = datos[3].split("(")[-1].split(")")[0]
    Visual=int(Estimulos.split("-")[0])
    Auditivo=float(Estimulos.split("-")[1])

    df_1 = df[df["Condicion"] == Condicion]
    df_2 = df_1[df_1["Camada"] == Camada]
    df_3 = df_2[df_2["Individuo"] == Individuo]
    df_4 = df_3[df_3["Visual"] == Visual]
    df_5 = df_4[df_4["Auditivo"] == Auditivo]
    
    #print(Camada, Individuo, Condicion, Visual, Auditivo)
    
    Respuesta=df_5["Respuesta"]
    Respuesta=int(Respuesta.iloc[0])
    
    print(Camada, Individuo, Condicion, Visual, Auditivo, Respuesta)
    
    if Respuesta==0:
        continue

    Frame=df_5["Frame"]
    Duracion=df_5["Duracion"]
    
    Frame=int(Frame.iloc[0])
    Duracion=int(Duracion.iloc[0])

    Trackingdf=pd.read_excel(Nombre)
    
    ventana = 18 # ventana temporal 

    t_inicial = Frame# - ventana # ventana temporal antes del evento
    t_final = Frame + ventana # ventana temporal después del evento (yo quiero que sea 20)

    cant_frames = ventana#*2 - 1

    evento = Trackingdf[(Trackingdf["bodyparts_coords"] > t_inicial) & (Trackingdf["bodyparts_coords"] < t_final)]
    
    #print(evento)
    
    if len(evento) != cant_frames:
        continue
    
    print("Evento válido.")
    
    #plt.plot(evento["bodyparts_coords"], evento["head_velocity"])
    #plt.show()
    
    curvaturemax = evento["sum_curvature"].max()
    
    if not "curvatura_maxima" in resultados[Condicion]:
        resultados[Condicion]["curvatura_maxima"] = []
    
    resultados[Condicion]["curvatura_maxima"].append(curvaturemax)
    
    sum_tail_angles = evento["sum_tail_angles"].max()
     
    if not "sum_tail_angles" in resultados[Condicion]:
        
        resultados[Condicion]["sum_tail_angles"] = []
    
    resultados[Condicion]["sum_tail_angles"].append(sum_tail_angles)
    
    sum_curvature = evento["sum_curvature"].max()
     
    if not "sum_curvature" in resultados[Condicion]:
        
        resultados[Condicion]["sum_curvature"] = []
    
    resultados[Condicion]["sum_curvature"].append(sum_curvature)
    
    if not "archivo" in resultados[Condicion]:
        
        resultados[Condicion]["archivo"] = []
    
    resultados[Condicion]["archivo"].append(Nombre)
    
    if not "frame" in resultados[Condicion]:
        
        resultados[Condicion]["frame"] = []
    
    resultados[Condicion]["frame"].append(Frame)
    
    eventos_por_condicion[Condicion] += 1
        
    
resultados
```

```python
for Condicion in resultados:
    for evento in range(eventos_por_condicion[Condicion]):
        archivo = resultados[Condicion]["archivo"][evento]
        frame = resultados[Condicion]["frame"][evento]
        sum_tail_angles = resultados[Condicion]["sum_tail_angles"][evento]
        
        print(archivo, frame, sum_tail_angles)
```
